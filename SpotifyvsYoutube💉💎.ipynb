# Sacando el IQR, MAX y Q1,Q3 üíÖ

```python
q1 = serieComments['25%']
q3 = serieComments['75%']
iqr = q3-q1
print(q1)
print(q3)
print(iqr)
```

```python
maxComen = q3 + 1.5*iqr
minComen = q1 - 1.5*iqr

print('El m√°ximo de nuestra distribuci√≥n es', maxComen)
print('El m√≠nimo de nuestra distribuci√≥n es', minComen)
```
