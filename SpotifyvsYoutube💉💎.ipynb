Columnas: 28

1. **Track**: nombre de la canción, tal como aparece en la plataforma de Spotify.
2. **Artista**: nombre del artista.
3. **Url_spotify:** la URL de la canción en Spotify.
4. **Álbum**: el álbum en el que se encuentra la canción en Spotify.
5. **Album_type**: indica si la canción se lanzó en Spotify como single o si está contenida en un álbum.
6. **Uri**: un enlace de Spotify utilizado para encontrar la canción a través de la API.
7. **Danceability**: describe cuán adecuada es una pista para bailar. Un valor de 0.0 es el menos bailable y 1.0 es el más bailable.
8. **Energy**: es una medida de 0.0 a 1.0 y representa una medida perceptual de intensidad y actividad.
9. **Key**: la nota músical en la que se encuentra la pista. Los enteros se asignan a los tonos utilizando la notación estándar de Pitch Class. Por ejemplo, 0 = C, 1 = C♯/D♭, 2 = D, y así sucesivamente. Si no se detectó ninguna clave, el valor es -1.
10. **Loudness**: el volumen general de una pista en decibelios (dB). Los valores de volumen se promedian en toda la pista y son útiles para comparar el volumen relativo de las pistas. El volumen es la calidad de un sonido que es el correlato psicológico primario de la fuerza física (amplitud). Los valores suelen oscilar entre -60 y 0 dB.
11. **Speechiness**: detecta la presencia de palabras habladas en una pista. Cuanto más exclusivamente parecido al habla sea la grabación (por ejemplo, programa de entrevistas, audiolibro, poesía), más cercano a 1.0 será el valor del atributo. Los valores por encima de 0.66 describen pistas que probablemente estén compuestas completamente por palabras habladas. Los valores entre 0.33 y 0.66 describen pistas que pueden contener tanto música como habla, ya sea en secciones o en capas, incluyendo casos como la música rap. Los valores por debajo de 0.33 representan probablemente música y otras pistas no parecidas al habla.
12. **Acousticness**: una medida de confianza de 0.0 a 1.0 de si la pista es acústica. 1.0 representa una alta confianza en que la pista es acústica.
13. **Instrumentalness**: predice si una pista no contiene voces.
14. **Liveness**: detecta la presencia de una audiencia en la grabación. Valores más altos de liveness representan una mayor probabilidad de que la pista se haya interpretado en vivo. Un valor por encima de 0.8 proporciona una fuerte probabilidad de que la pista sea en vivo.
15. **Valence**: una medida de 0.0 a 1.0 que describe la positividad musical transmitida por una pista. Las pistas con alta valencia suenan más positivas (por ejemplo, felices, alegres, eufóricas), mientras que las pistas con baja valencia suenan más negativas (por ejemplo, tristes, deprimidas, enojadas).
16. **Tempo**: el tempo estimado general de una pista en pulsaciones por minuto (BPM). En términos musicales, el tempo es la velocidad o ritmo de una pieza determinada y se deriva directamente de la duración promedio del pulso.
Duration_ms: la duración de la pista en milisegundos.
17. **Stream**: número de reproducciones de la canción en Spotify.
18. **Url_youtube**: la URL del video vinculado a la canción en Youtube, si existe alguno.
19. **Title**: título del videoclip en youtube.
20. **Channel**: nombre del canal que ha publicado el video.
21. **Views**: número de vistas.
22. **Likes**: número de "me gusta".
23. **Comments**: número de comentarios.
24. **Description**: descripción del video en Youtube.
25. **Licensed**: indica si el video representa contenido con licencia, lo que significa que el contenido fue cargado en un canal vinculado a un socio de contenido de YouTube y luego reclamado por ese socio.
26. **official_video**: valor booleano que indica si el video encontrado es el video oficial de la canción.
27. **#**: el orden
28. **# Stream**: codigo del stream

# Sacando el IQR, MAX y Q1,Q3 💅

```python
q1 = serieComments['25%']
q3 = serieComments['75%']
iqr = q3-q1
print(q1)
print(q3)
print(iqr)
```

```python
maxComen = q3 + 1.5*iqr
minComen = q1 - 1.5*iqr

print('El máximo de nuestra distribución es', maxComen)
print('El mínimo de nuestra distribución es', minComen)
```
